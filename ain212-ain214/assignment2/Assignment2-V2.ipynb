Question 5) Create a new column named as 'happiness_degree' which labels value of 'ladder'
*very happy if ladder < 30
*happy if 30 <= ladder < 60
*moderately happy if 60 <= ladder < 90
*unhappy if 90 <= ladder < 120
*very unhappy if ladder >= 120
Print the top 10 records updated dataframe (10 points)

```python
#codes
happiness_df["happiness_degree"]=""
for i in range(len(happiness_df)) :
      
    if happiness_df.iloc[i][1] <30 :
        happiness_df.iloc[[i],[-1]]="very happy"
    
    elif happiness_df.iloc[i][1] <60:
        happiness_df.iloc[[i],[-1]]="happy"
    
    elif happiness_df.iloc[i][1]<90 :
        happiness_df.iloc[[i],[-1]]="moderately happy"
    
    elif happiness_df.iloc[i][1]<120 :
        happiness_df.iloc[[i],[-1]]="unhappy"
    
    else:
        happiness_df.iloc[[i],[-1]]="very unhappy"
        
happiness_df.head(10)
```

Question 10) Draw random samples 30%, 40%, 50%, 60% of the data with replacement and investigate how the correlation of sample to the whole data changes with respect to sample size. (15 points)

```python
#codes
length=len(happiness_df)

x30sample_df=happiness_df.sample(n=int(length*0.3),replace=True,random_state=1)
x40sample_df=happiness_df.sample(n=int(length*0.4),replace=True , random_state=1)
x50sample_df=happiness_df.sample(n=int(length*0.5),replace=True ,random_state=1)
x60sample_df=happiness_df.sample(n=int(length*0.6),replace=True ,random_state=1)
"""The higher the percentage of samples, the more the correlation table of sample looks like the correlation table of our data. 
We can easily say that while looking the tables. For example, lets look at corr between Corruption and Ladder. 
In original sample, it is 0.224394, when we look at the %30 of sample it is 0.326008, %40 0.244066, %50 0.167272, %60 0.239145.
As we see. The most accurate result in %60 of sample. This theorem can not be always true but generally it is true."""
x30sample_df.corr()
```
